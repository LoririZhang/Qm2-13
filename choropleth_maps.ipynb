```python
# Creating a dataframe for google trends

google = pandas.read_csv('./data/Qm2_13/google_data.csv')
google_data = '/content/data/Qm2_13/goole_data.csv'
google.head()
```

```python
#Reordering the google data alphabetically

df1 = google.sort_values('Region')
df1
```

```python
#joining the two dataframes and creating a table with Google Trends and State Code

gsc = df1.merge(statecode, left_index=True, right_index=True)
gsc
```

```python
#@title Visualisation: Choropleth map

import plotly.express as px
import pandas as pd

gsc=gsc[['Region', 'State code','ResearchTrends']]
gsc.rename({'ResearchTrends': 'Google Search Trends for climate change'}, axis=1, inplace=True)
gsc.rename({'Region': 'State'}, axis=1, inplace=True)


gsc.head()
```

```python
fig = px.choropleth (gsc,
                     locations='State code',
                     locationmode="USA-states",
                     scope="usa",
                     color='Google Search Trends for climate change',
                     color_continuous_scale="Viridis_r",
                     
                     )

fig.show()
```

```python
fig.update_layout(
    title_text = 'Google Searches for the term climate change in 2021 by State',
    title_font_family="Times New Roman",
    title_font_color="black",
    title_x=0.45,
      )
```
